```python
import matplotlib.pyplot as plt #importing matplot libary
import pandas as pd #importing pandas libary

#Average Admission grade per Marital status
avg_grade = df.groupby('Marital status')['Admission grade'].mean()

#Extract the index of the average grade and convert into list
marital_status = list(avg_grade.index)

#extract the values of the average grade
avg_admission_grade = avg_grade.values

# Plotting bar chart
plt.figure(figsize=(12, 4))
plt.bar(marital_status, avg_admission_grade, color='blue')

# Labels and title
plt.xlabel('Marital Status')
plt.ylabel('Average Admission Grade')
plt.title('Average Admission Grade per Marital Status')

# Rotate x-axis labels 
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
# Group by Course and get average Admission grade
avg_grade = df.groupby('Course')['Admission grade'].mean()

# Categories and values
courses = list(avg_grade.index)
avg_grades = list(avg_grade.values)

# Plot
plt.figure(figsize=(12, 5))
plt.bar(range(len(courses)), avg_grades, color='blue')
plt.title('Average Admission Grade per Course')
plt.xlabel('Course')
plt.ylabel('Average Admission Grade')
plt.xticks(range(len(courses)), courses, rotation=90)
plt.tight_layout()
#plot to show the diagram
plt.show()
```

```python
import pandas as pd #import all the necessary libaries
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
print(df.columns) #print all the columns
```

```python
print(df['Target'].unique()) #print the target
```

```python
y = df['Target']  

# Features
x = df.drop(columns='Target')
```

```python
x = pd.get_dummies(x, drop_first=True) #dropping first to remove redundancy
```

```python
from sklearn.model_selection import train_test_split # importing the train_test_split

#Split the data into training (70%) and testing (30%) sets

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
decisiontree = DecisionTreeClassifier(max_depth=5, random_state=42) #create a decisiontreeclassifier
```

```python
decisiontree.fit(x_train, y_train) #train the decisiontreemodel
```

```python
y_pred_decisiontree = decisiontree.predict(x_test) #predictions on the test set
```

```python
print(classification_report(y_test, y_pred_decisiontree)) #print classification report
```

```python
from sklearn.metrics import confusion_matrix #import confusion_matrix

cm = confusion_matrix(y_test, y_pred_decisiontree) 
print("Confusion Matrix:\n", cm) #print
```

```python
from sklearn.metrics import cohen_kappa_score #import cohen_kappa_score

#calculating the predictions by cohen_kappa_score

kappa = cohen_kappa_score(y_test, y_pred_decisiontree)
print("Cohen's Kappa:", kappa) #print cohen's kappa
```

```python
from sklearn.metrics import matthews_corrcoef # import matthews_corrcoef

#predictions by matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred_decisiontree)
print("Matthews Correlation Coefficient:", mcc)
```

```python
#feature importance
feature_importance = pd.Series(decisiontree.feature_importances_, index=x.columns).sort_values(ascending=False)
#plotting figure size
plt.figure(figsize=(10,3))
#plotting bar and title
feature_importance.plot(kind='bar', title='Feature Importance (Decision Tree)')
#plotting label
plt.ylabel('Importance')

#plot to show
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=5) #create a KNeighborsClassifier
```

```python
knn.fit(x_train, y_train) #training data
```

```python
#Use the trained KNN model to predict the outcomes for the test dataset
y_pred_knn = knn.predict(x_test) 
```

```python
print(classification_report(y_test, y_pred_knn)) #prepare classification report
```

```python
from sklearn.metrics import cohen_kappa_score #importing cohen_kappa_score

# Using Knn for predictions
kappa_knn = cohen_kappa_score(y_test, y_pred_knn)
print("Cohen's Kappa:", kappa_knn)
```

```python
from sklearn.metrics import confusion_matrix #importing confusion_matrix

#using confusion matrix for predictions
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix:\n", cm_knn)
```

```python
from sklearn.metrics import balanced_accuracy_score #importing balanced_accuracy_score
#using balanced_accuracy_score
balanced_acc = balanced_accuracy_score(y_test, y_pred_knn)
#printing the score
print("Balanced Accuracy:", balanced_acc)
```

```python
from sklearn.inspection import permutation_importance #importing permutation_importance
```

```python
result = permutation_importance(knn, x_test, y_test, n_repeats=10, random_state=42, scoring='accuracy')#checking the results
```

```python
#convert feature importance into list
feature_importance_knn = pd.Series(result.importances_mean, index=x.columns).sort_values(ascending=False)
```

```python
plt.figure(figsize=(7,3)) #Figuresize
feature_importance_knn.plot(kind='bar', title='Feature Importance (KNN)') #plot title
plt.ylabel('Mean Decrease in Accuracy') #plot label
plt.show() #plot to show
```

```python
from sklearn.linear_model import LogisticRegression #importing logistic regression
#Split data into 70% train and 30% test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#Training the logistic regresssion model
logistic_regression = LogisticRegression(max_iter=50)
logistic_regression.fit(x_train, y_train)
```

```python
y_pred = logistic_regression.predict(x_test)#Predict labels for the test set
```

```python
print(classification_report(y_test, y_pred))#printing the classification report
```

```python
from sklearn.metrics import cohen_kappa_score # importing cohen_kappa_score
from sklearn.metrics import confusion_matrix #importing confusion_matrix

# Measures predictions and actual labels
confusionmatrix_lr = confusion_matrix(y_test, y_pred)
kappa_lr = cohen_kappa_score(y_test, y_pred)

#print confusion matric and cohen's kappa results
print("Confusion Matrix:\n", confusionmatrix_lr)
print("Cohen's Kappa:", kappa_lr)
```

```python
from sklearn.metrics import log_loss #importing log_loss
y_prob = logistic_regression.predict_proba(x_test)[:,1]
# Log Loss for logistic regression
logloss_logisticregression = log_loss((y_test=='Graduate').astype(int), y_prob)
print("Log Loss:", logloss_logisticregression)
```

```python
#Feature importance for logistic regression
feature_importance_logisticregression = pd.Series(abs(logistic_regression.coef_[0]), index=x.columns).sort_values(ascending=False)
```

```python
#plotting figuresize
plt.figure(figsize=(8,1))
#plotting title 
feature_importance_lr.plot(kind='bar', title='Feature Importance (Logistic Regression)')
#plotting label
plt.ylabel('Importance Score')
#plot to show
plt.show()
```
